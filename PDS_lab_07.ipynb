##Question-01

```python
import pandas as pd
import matplotlib.pyplot as plt

weather_df = pd.read_csv('/content/daily_temperature.csv')

plt.figure(figsize=(10, 5))
plt.plot(weather_df['Day'], weather_df['Temperature'], color='blue', linewidth=2, marker='o')

plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel('Day of Month')
plt.ylabel('Temperature (°C)')
plt.title('Daily Temperature Changes Over a Month')

plt.tight_layout()
plt.show()

max_temp_day = weather_df.loc[weather_df['Temperature'].idxmax()]
min_temp_day = weather_df.loc[weather_df['Temperature'].idxmin()]

print(f"Hottest Day: Day {max_temp_day['Day']} with {max_temp_day['Temperature']}°C")
print(f"Coldest Day: Day {min_temp_day['Day']} with {min_temp_day['Temperature']}°C")
```

##Question-02

```python
import pandas as pd
import matplotlib.pyplot as plt

sales_df = pd.read_csv('/content/sales_data.csv')

plt.figure(figsize=(8, 5))
plt.plot(sales_df['Quarter'], sales_df['Sales'], color='green', linewidth=2, marker='o', label='Sales')

peak_index = sales_df['Sales'].idxmax()
plt.scatter(sales_df.loc[peak_index, 'Quarter'],
            sales_df.loc[peak_index, 'Sales'],
            color='red', s=100, label='Peak Sales')

plt.grid(True, linestyle='--', alpha=0.6)
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Quarterly Sales Performance')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Peak Sales occurred in {sales_df.loc[peak_index, 'Quarter']} "
      f"with sales of {sales_df.loc[peak_index, 'Sales']} units.")

if len(sales_df) >= 4:
    print("\nObservation: Sales tend to fluctuate by quarter; peaks may correspond to festive seasons or year-end demand.")
```
